## Using Range:

```python
def range_function(dataset,start,end):
    from scipy.stats import norm
    import seaborn as sns
    import matplotlib.pyplot as plt
    sns.distplot(dataset,kde=True,kde_kws={'color':"blue"},color='Orange')
    plt.axvline(start,color='Green')
    plt.axvline(end,color='Red')
    # sample
    # sample=dataset
    # sample_mean=sample.mean()
    # sample_std=sample.std()
    mean=dataset.mean()
    std=dataset.std()
    print(f"Mean:{mean}\nStd:{std}")
    dist=norm(mean,std)
    values=[value for value in range(start,end)]
    probability=[dist.pdf(value) for value in values]
    sums_prob=sum(probability)
    
    print(f"Range({start},{end})----> pdf : {sums_prob}")
    
```

```python
range_function(load['ssc_p'],70,100)
```
