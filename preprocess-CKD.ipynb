## find missing value percentage:

```python
a=(load.isnull().sum() / len(load))*100

# <5 ---> simple fillna
#  5< and <20  ---> middle level
# >20  ---->advanced level handling
```

## findout categorical and numerical column:

```python
# reason --->1.cate ku mode use pannalam and 2.num ku median or mean use panni missing value aa handle pannalam 
def can_num(data):
    cate=[]
    num=[]
    for column in data.columns:
        if data[column].dtypes =='O':
            cate.append(column)
        else:
            num.append(column)
    return cate,num
```

```python
cate,num=can_num(load)
```

```python
num
```

## To separate the column for levels(low,medium,high (null values)):

```python
low=[]
medium=[]
high=[]
for column in load.columns:
    values=a[column]
    if 0< values <=5:
        low.append(column)
    elif 5< values <20:
        medium.append(column)
    elif values >=20:
        high.append(column)
        
```

```python
low
```

```python
medium
```

```python
high
```

```python
low_cate=[]
low_num=[]
for column in low:
    if load[column].dtypes=='O':
        low_cate.append(column)
    else:
        low_num.append(column)
```

```python
low_cate
```

```python
low_num
```

```python
# use median value (simple method):
for column in low_num:
    load[column]=load[column].fillna(load[column].median())
```

```python
load.isna().sum()
```

```python
# categorical value use model(most frequency value(simple method)):
for column in low_cate:
    load[column]=load[column].fillna(load[column].mode()[0])
```

```python
load.isna().sum()
```

```python
# diff mid level :
mid_cate=[]
mid_num=[]
for column in medium:
    if load[column].dtypes=='O':
        mid_cate.append(column)
    else:
        mid_num.append(column)
```

```python
mid_cate
```

```python
load['pc']=load["pc"].fillna(load["pc"].mode()[0])
```

```python
mid_num
```

```python
for i in mid_num:
    load[i]=load[i].fillna(load[i].median())
```
