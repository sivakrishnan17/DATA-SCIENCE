```python
# ṣeparate quan and qual:

quan=[]
qual=[]
def quan_qual(data):
    for i in data.columns:
        if data[i].dtypes=='str':
            qual.append(i)
        else:
            quan.append(i)
    return quan,qual
print(quan_qual(data))
```

### Data Frame :

```python
df=pd.DataFrame(index=['mean','median','mode'],columns=quan)
```

```python
for column_name in quan:
    df.loc["mean",column_name]=data[column_name].mean()
    df.loc["median",column_name]=data[column_name].median()
    df.loc["mode",column_name]=data[column_name].mode()[0]
    
```
