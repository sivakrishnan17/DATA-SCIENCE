## Outlier finding and filter phase:

```python
df=pd.DataFrame(index=['mean','median','mode',"25%","75%","max","IQR","MUL*1.5","upper_outlier","lower_outlier","min","maxx","var","std"],columns=quan)
```

```python
for column_name in quan:
    df.loc["mean",column_name,]=data[column_name].mean()
    df.loc["median",column_name]=data[column_name].median()
    df.loc["mode",column_name]=data[column_name].mode()[0]
    df.loc["25%",column_name]=np.percentile(data[column_name],25)
    df.loc["75%",column_name]=np.percentile(data[column_name],75)
    df.loc["max",column_name]=np.percentile(data[column_name],100)
    df.loc['IQR',column_name]=(df.loc["75%",column_name])-(df.loc["25%",column_name])
    df.loc["MUL*1.5",column_name]=1.5 * ( df.loc['IQR',column_name])
    df.loc["upper_outlier",column_name]=(df.loc["75%",column_name])+df.loc["MUL*1.5",column_name]
    df.loc["lower_outlier",column_name]=(df.loc["25%",column_name]-df.loc["MUL*1.5",column_name])
    df.loc["min",column_name]=data[column_name].min()
    df.loc["maaxx",column_name]=data[column_name].max()
    # var and std
    df.loc["var",column_name]=data[column_name].var()
    df.loc["std",column_name]=data[column_name].std()
    
```
