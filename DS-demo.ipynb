## 2.Data Frame:

```python
# mean,mode,median
# percentile(k\100(n+1))
#IQR and outliers
df=pd.DataFrame(index=["mean","median","mode","q1","q3","IQR","mul*1.5","min","less_range","max","grater_range"],columns=quan)
```

```python
df
```

```python
def quan_qual(data,quan):
    for column in quan:
        df.loc["mean",column]=data[column].mean()
        df.loc["median",column]=data[column].median()
        df.loc["mode",column]=data[column].mode()[0]
        # percentile:
        df.loc["q1",column]=np.percentile(data[column],25)
        df.loc["q3",column]=np.percentile(data[column],75)
        df.loc["min",column]=data[column].min()
        df.loc["max",column]=data[column].max()
        
        # IQR(interquatile range):
        df.loc["IQR",column]= df.loc["q3",column] - df.loc["q1",column]
        df.loc["mul*1.5",column]=1.5 * (df.loc["IQR",column])
        df.loc["less_range",column]=(df.loc["q1",column])-(df.loc["mul*1.5",column])
        df.loc["grater_range",column]= (df.loc["q3",column])+(df.loc["mul*1.5",column]) 
```

## Outlier finding:

```python
lesser=[]
grater=[]
for column in quan:
    if df.loc["less_range",column] > df.loc["min",column]:
        lesser.append(column)
    if df.loc["grater_range",column] < df.loc["max",column]:
        grater.append(column)
    
```

```python
print("lesser:",lesser)
print("grater :",grater)
```

## outlier replace method:
## to replace in out original dataset not dataframe:


```python
for column in lesser:
    data.loc[data[column]<df[column]["less_range"],column]=df[column]["less_range"]
```

```python
for column in grater:
    data.loc[data[column]>df[column]["grater_range"],column]=df[column]["grater_range"]
```

## Frequecy :

### Create a frequency table:

```python
frequency_df=pd.DataFrame(columns=["value","frequency","relative frequecy(%)","cumsum"])
```

```python
# frequency_df.drop("hsc_p",axis=1)
```

```python
frequency_df.columns
```

```python
def Frequency(data,column):
    frequency_df["value"]= data[column].value_counts().index
    frequency_df["frequency"]= data[column].value_counts().values
    frequency_df["relative frequecy(%)"]=(frequency_df["frequency"])/91
    frequency_df["cumsum"]=frequency_df["relative frequecy(%)"].cumsum()
```
